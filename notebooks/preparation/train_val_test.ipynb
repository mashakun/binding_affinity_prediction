```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

### Splitting

```python
df_sm_fp = pd.read_csv('/content/drive/MyDrive/Course_work/notebooks/data/sm_fp_df.csv')

print(df_sm_fp.shape)
```

```python
y = df_sm_fp.affinity
x = df_sm_fp.drop(columns = ['affinity']).copy()
```

```python
x_train, x_rem, y_train, y_rem = train_test_split(x, y, train_size=0.7)
x_val, x_test, y_val, y_test = train_test_split(x_rem, y_rem, test_size=0.33)
```

```python
code_train = x_train.pdbcode.values
code_val = x_val.pdbcode.values
code_test = x_test.pdbcode.values

train_affinity = y_train.values
val_affinity = y_val.values
test_affinity = y_test.values
```

### Creating splitted datasets

```python
len(test_affinity)
```

```python
df_train = pd.DataFrame({'pdbcode' : code_train, 'affinity': train_affinity})
df_val = pd.DataFrame({'pdbcode' : code_val, 'affinity': val_affinity})
df_test = pd.DataFrame({'pdbcode' : code_test, 'affinity': test_affinity})
```

```python
print(df_train.shape, df_val.shape, df_test.shape)
```

```python
df_train.to_csv('/content/drive/MyDrive/Course_work/notebooks/data/set_train.csv')
df_val.to_csv('/content/drive/MyDrive/Course_work/notebooks/data/set_val.csv')
df_test.to_csv('/content/drive/MyDrive/Course_work/notebooks/data/set_test.csv')
```
